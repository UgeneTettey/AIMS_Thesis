```python
#importing relevant libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

### Simulations for each compartment of the model

```python
#--------------------------
# Assumed parameter values
#--------------------------
Lambda = 0.35  #knust unirank
mu = 0.25 #[32]
tau1 = 0.1 #est
psi = 0.14 #0.03
delta = 0.35 #[32]
tau = 0.2 #estimated
alpha1 = 0.20 #est
rho = 0.23 #est
sigma = 0.3 #[1]
B = 0.2 #[32]
beta1 = 0.7 #sourced [1]
beta2 = 0.8
beta3 = 0.5 #0.07
phi = 0.1
psi2 = 0.8
theta = 0.06


#--------------------------------------------------------
# Function to define the system of differential equations
#--------------------------------------------------------
def abuse_model(y, t):
    S, DP, DO, DA, RC, R = y

    dS_dt = Lambda + tau * R - (beta1 * DP + beta2 * DO + beta3 * DA + mu) * S
    dDP_dt = beta1 * S * DP - (theta + sigma + phi + mu + tau1) * DP
    dDO_dt = beta2 * S * DO + theta * DP - (alpha1 + psi + mu + psi2) * DO
    dDA_dt = beta3 * S * DA + psi2 * DO - (delta + mu + rho) * DA
    dRC_dt = sigma * DP + alpha1 * DO + rho * DA - (mu + B) * RC
    dR_dt = B * RC + phi * DP - (mu + tau) * R

    return [dS_dt, dDP_dt, dDO_dt, dDA_dt, dRC_dt, dR_dt]


#--------------------
# Initial conditions
#--------------------
S0 = 0.3
DP0 = 0.22
DO0 = 0.2
DA0 = 0.15
RC0 = 0.08
R0 = 0.05
initial_conditions = [S0, DP0, DO0, DA0, RC0, R0]

# Time points
months = 48   #simulating over 4 years
time_points = 1000
t = np.linspace(0, months, time_points)

# Solve the system of differential equations
sol = odeint(abuse_model, initial_conditions, t)

# Plot the results for each compartment
plt.figure(figsize=(10, 8))

plt.subplot(3, 2, 1)
plt.plot(t, sol[:, 0] , 'orange')
plt.xlabel('Time (months)')
plt.ylabel('Population')
plt.title('Susceptible Students (S)')

plt.subplot(3, 2, 2)
plt.plot(t, sol[:, 1], color='blue')
plt.xlabel('Time (months)')
plt.ylabel('Population')
plt.title( r'Drug Users by Peer Influence ($D_P$)')

plt.subplot(3, 2, 3)
plt.plot(t, sol[:, 2], 'black')
plt.xlabel('Time (months)')
plt.ylabel('Population')
plt.title(r'Occasional Drug Users ($D_O$)')

plt.subplot(3, 2, 4)
plt.plot(t, sol[:, 3], 'red')
plt.xlabel('Time (months)')
plt.ylabel('Population')
plt.title(r'Drug Addicts ($D_A$)')

plt.subplot(3, 2, 5)
plt.plot(t, sol[:, 4], 'green')
plt.xlabel('Time (months)')
plt.ylabel('Population')
plt.title(r'Students Under Rehab and Counselling ($R_C$)')

plt.subplot(3, 2, 6)
plt.plot(t, sol[:, 5], )
plt.xlabel('Time (months)')
plt.ylabel('Population')
plt.title('Recovered Students (R)')

plt.tight_layout()
plt.show()
```

##### The  code simulates the system of equations representing substance use and abuse among university students and generates plots to visualize the behavior of the variables over time. Let's discuss the outputs of the graphs:

 #### S (Susceptible Students):
This plot shows the population of students who are susceptible to drug use and abuse (S) over time.
Initially, the population of susceptible students may decrease due to interactions with other drug-using students (DP, DO, DA) and natural deaths (mu).
The recruitment rate (Lambda) and the rate at which recovered students become susceptible again (tau) contribute to the increase in the susceptible population.The plot helps monitor the fluctuations in the susceptible student population throughout the simulation.

#### DP (Students using drugs by peer influence):
This plot represents the population of students who use and abuse drugs due to peer influence (DP) over time.
The DP population is influenced by the interaction between susceptible students (S) and DP, as well as the recovery rate from the DP compartment (phi).The drug-induced death rate in the DP compartment (tau1) and other death rates (mu) contribute to the decrease in the DP population. The plot provides insights into the dynamics of drug use influenced by peer interactions.

####  DO (Students occasionally using drugs):
This plot shows the population of students who occasionally use drugs (DO) over time. 
The DO population is influenced by the interaction between susceptible students (S) and DO, as well as the inflow from the DP compartment (theta).
The drug-induced death rate in the DO compartment (alpha1), other death rates (mu), and the rate at which students in DO visit rehabilitation centers (psi) impact the DO population.
The plot helps visualize the transitions and fluctuations in occasional drug use among students.

####    DA (Students addicted to drugs):
This plot represents the population of students who are addicted to drugs (DA) over time.
The DA population is influenced by the interaction between susceptible students (S) and DA, as well as the inflow from the DO compartment (psi2).
The drug-induced death rate in the DA compartment (delta), other death rates (mu), and the rate at which students in DA visit rehabilitation centers (rho) affect the DA population.
The plot illustrates the changes in the population of students struggling with drug addiction.

####    RC (Students under rehab and counseling):
This plot shows the population of students receiving rehabilitation and counseling (RC) over time.
The RC population is influenced by the inflow from the DP, DO, and DA compartments (sigma, alpha1, rho) and the recovery rate from rehabilitation centers (B).
The natural death rate (mu) and other death rates impact the RC population.
The plot provides insights into the population of students seeking and undergoing rehabilitation.

####   R (Recovered Students):
This plot represents the population of students who have recovered from drug use and no longer use drugs (R) over time.
The R population is influenced by the recovery rate from rehabilitation centers (B) and the inflow from the DP compartment (phi).
The natural death rate (mu) and the rate at which recovered students become susceptible again (tau) impact the R population.
The plot helps monitor the transitions and fluctuations in the population of students who have successfully recovered from drug addiction.

## Performing sensitivity Analysis

### Sensitivity on R_0

```python
Lambda = 0.35  #sourced
mu = 0.25  #sourced
psi = 0.14
psi_2 = 0.8
alpha1 = 0.20  #est
beta_2 = 0.8  #assumed

R_0 = (beta_2 * Lambda * (alpha1 + psi + psi_2 + mu)) / mu
def sensitivity_index(beta_2, Lambda, alpha_1, psi, psi_2, mu):
    dR_0_dbeta_2 = (Lambda * (alpha_1 + psi + psi_2 + mu)) / mu
    dR_0_dLambda = (beta_2 * (alpha_1 + psi + psi_2 + mu)) / mu
    dR_0_dalpha_1 = (beta_2 * Lambda) / mu
    dR_0_dpsi = (beta_2 * Lambda) / mu
    dR_0_dpsi_2 = (beta_2 * Lambda) / mu
    dR_0_dmu = (-beta_2 * Lambda * (alpha_1 + psi + psi_2)) / (mu ** 2)

    sensitivity_dict = {
        "beta_2": dR_0_dbeta_2 * (beta_2 / R_0),
        "Lambda": dR_0_dLambda * (Lambda / R_0),
        "alpha_1": dR_0_dalpha_1 * (alpha_1 / R_0),
        "psi": dR_0_dpsi * (psi / R_0),
        "psi_2": dR_0_dpsi_2 * (psi_2 / R_0),
        "mu": dR_0_dmu * (mu / R_0),
    }

    return sensitivity_dict


#=========================
print("Calculating for sensitivity indices: \n")


# Calculate the sensitivity indices
sensitivity_indices = sensitivity_index(beta_2, Lambda, alpha1, psi, psi_2, mu)

# Print the sensitivity indices
for parameter, sensitivity_index in sensitivity_indices.items():
    print(f"Sensitivity index ({parameter}): {sensitivity_index}")
```
